```python
# Install necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style for better visualization
sns.set(style="whitegrid")

# Step 1: Data Preparation & Loading
# Simulating patient blood pressure values between 100 and 180 mmHg (with variability)
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=140, scale=15, size=2000)  # Mean=140, SD=15

# Convert to a pandas DataFrame
df = pd.DataFrame(data, columns=["Systolic_BP"])

# Show the first few rows of the DataFrame
df.head()
```

```python
# Plot the histogram of the full dataset to visualize the blood pressure distribution
plt.figure(figsize=(8, 6))
plt.hist(df["Systolic_BP"], bins=30, color='skyblue', edgecolor='black')
plt.title("Blood Pressure Distribution of 2,000 Patients")
plt.xlabel("Systolic Blood Pressure (mmHg)")
plt.ylabel("Frequency")
plt.show()

# Function to simulate random samples and compute sample means
def simulate_sampling(sample_size, num_samples=1000):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(df["Systolic_BP"], size=sample_size, replace=False)
        sample_means.append(np.mean(sample))
    return sample_means
```

```python
# Step 3: Sampling and Visualization for Different Sample Sizes
sample_sizes = [25, 50, 100]
sample_means_dict = {}

# Generate sample means for each sample size
for size in sample_sizes:
    sample_means_dict[size] = simulate_sampling(size)

# Plot histograms of the sample means for each sample size
plt.figure(figsize=(12, 8))
for i, size in enumerate(sample_sizes, 1):
    plt.subplot(2, 2, i)
    plt.hist(sample_means_dict[size], bins=30, color='skyblue', edgecolor='black')
    plt.title(f"Sample Means for Sample Size {size}")
    plt.xlabel("Mean Systolic Blood Pressure (mmHg)")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Step 4: Statistical Insight

# Compute and display the mean, median, variance, and standard deviation of the full dataset
full_mean = np.mean(df["Systolic_BP"])
full_median = np.median(df["Systolic_BP"])
full_variance = np.var(df["Systolic_BP"])
full_std = np.std(df["Systolic_BP"])

print(f"Full Dataset Statistics:")
print(f"Mean: {full_mean:.2f} mmHg")
print(f"Median: {full_median:.2f} mmHg")
print(f"Variance: {full_variance:.2f}")
print(f"Standard Deviation: {full_std:.2f} mmHg")

# Compare how the distribution of sample means narrows as the sample size increases
plt.figure(figsize=(10, 6))
for size in sample_sizes:
    plt.hist(sample_means_dict[size], bins=30, alpha=0.5, label=f"Sample Size {size}")
plt.title("Distribution of Sample Means for Different Sample Sizes")
plt.xlabel("Mean Systolic Blood Pressure (mmHg)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Step 5: Explain relationship between sample size and accuracy
# A large sample size leads to a more precise estimate of the population mean.
# As the sample size increases, the variability of the sample means decreases.
# Step 6: Adaptation: Allowing User to Set a Custom Sample Size

# Function to plot standard error vs. sample size
def plot_standard_error_vs_sample_size():
    sample_sizes_custom = np.arange(10, 200, 10)
    standard_errors = []

    for size in sample_sizes_custom:
        sample_means = simulate_sampling(size)
        standard_errors.append(np.std(sample_means))

    plt.figure(figsize=(10, 6))
    plt.plot(sample_sizes_custom, standard_errors, marker='o', linestyle='-', color='purple')
    plt.title("Standard Error vs. Sample Size")
    plt.xlabel("Sample Size")
    plt.ylabel("Standard Error")
    plt.grid(True)
    plt.show()

# Plot standard error vs. sample size
plot_standard_error_vs_sample_size()
```
